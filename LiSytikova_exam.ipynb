# Экзаменационное задание

```python
with open('/content/shakespeare.txt', 'r') as file:
    text = file.read()
    print(text)
```

```python
def _tokenize(text):
        tokens = text.lower().split()
        return tokens

tokens = _tokenize(text)
tokens
```

```python
def remove_stopwords(tokens):
  stop_words = stopwords.words('english')
  tokens_chistye = [i for i in tokens if i.lower() not in stop_words]
  return tokens_chistye


tokens_chistye = remove_stopwords(tokens)
tokens_chistye
```

```python
def _analyze_word_frequencies(tokens_chistye):
   word_frequencies = Counter(tokens_chistye)
   return word_frequencies

word_frequencies = _analyze_word_frequencies(tokens_chistye)
word_frequencies
```

```python
def word_frequency_analysis(file_path, n=5):
    most_common_words = word_frequencies.most_common(n)
    return most_common_words
```

```python
file_path = '/content/shakespeare.txt'
most_common = word_frequency_analysis(file_path, 10)
most_common
```
